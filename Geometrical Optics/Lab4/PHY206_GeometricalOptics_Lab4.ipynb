#### Part 1: Normal Vision.

##### We begin with an eye model with normal vision, where the distance from the cornea to the retina was measured to be $d_i=10.8$cm.  The image was sharply focused on the retina for an object distance of $d_o=37$cm, so the calculated focal length of this eye model was $f=8.36$cm.

```python
'''Normal Vision'''
focal_length_eye = 8.36  #measured focal length of eye model [cm]
object_distance = 37     #measured object distance of eye model [cm]

'''Draw ray diagram'''
normal = geometrical_optics()
normal.add_lens(focal_length_eye, object_distance)
fig, ax1, ax2 = normal.ray_diagram()
```

```python
'''Add retina to ray diagram'''
retina = ConnectionPatch((normal.img_dict[1][2], 0.5), (normal.img_dict[1][2], -0.5), coordsA = 'data',
                         coordsB = 'data', linestyle = 'solid', color = 'black')
ax1.add_patch(retina)
ax1.annotate('Normal\nRetina', xy = (normal.img_dict[1][2], 0.6), ha = 'center')

'''Show new figure'''
fig
```

#### Part 2: Hyperopic Vision.

##### The retina of the eye model was then placed in the near position so that the distance from the cornea to the retina was measured to be $d_i=8.8$cm.  The image was then formed behind the retina, so we added a converging lens of focal length $f=40$cm to correct the hyperopic vision.

```python
'''Hyperopic Vision'''
focal_length_eye = 8.36  #measured focal length of eye model [cm]
object_distance = 37     #measured object distance of eye model [cm]

'''Draw ray diagram'''
hyperopia = geometrical_optics()
hyperopia.add_lens(focal_length_eye, object_distance)
fig, ax1, ax2 = hyperopia.ray_diagram()
```

```python
'''Add retina to ray diagram'''
hyperopia_retina = ConnectionPatch((normal.img_dict[1][2] - 2, 0.5), (normal.img_dict[1][2] - 2, -0.5), coordsA = 'data',
                         coordsB = 'data', linestyle = 'solid', color = 'black')
ax1.add_patch(hyperopia_retina)
ax1.annotate('Hyperopic\nRetina', xy = (normal.img_dict[1][2] - 2, 0.6), ha = 'center')

'''Show new figure'''
fig
```

```python
fig.savefig('c:/Users/Tucker Knaak/Downloads/HyperopicVision.png')
```

##### Add a convering lens with focal length $f=40$cm in front of the cornea to make the light converge at a smaller distance.

```python
hyperopia_corrected = geometrical_optics()
hyperopia_corrected.add_lens(40, 37)
hyperopia_corrected.add_lens(8.36, 495)
fig, ax1, ax2 = hyperopia_corrected.ray_diagram()
```

```python
'''Add retina to ray diagram'''
hyperopia_retina = ConnectionPatch((normal.img_dict[1][2] - 1, 2.5), (normal.img_dict[1][2] - 1, -2.5), coordsA = 'data',
                         coordsB = 'data', linestyle = 'solid', color = 'black')
ax1.add_patch(hyperopia_retina)
ax1.annotate('Hyperopic\nRetina', xy = (normal.img_dict[1][2] + 23, -3.5), ha = 'center')

'''Show new figure'''
fig
```
