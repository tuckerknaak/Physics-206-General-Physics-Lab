#### Part 1: $f_0=10$cm and $f_e=20$cm.

##### A proper microscope has an objective lens with a small focal length and an eyepiece lens with a larger focal length.  If the separation of the lenses is comparable to $L\approx f_0+f_e$, then the students will have created a compound microscope.

```python
'''do1 < fo, do2 < fe; Virtual, upright, enlarged final image --> no microscope'''
case1 = geometrical_optics()
case1.add_lens(10, 5)
case1.add_lens(20, 15)
case1.ray_diagram(save_diagram = True)
```

```python
'''do1 < fo, do2 > fe; Real, inverted, enlarged final image --> no microscope'''
case2 = geometrical_optics()
case2.add_lens(10, 5)
case2.add_lens(20, 45)
case2.ray_diagram(save_diagram = True)
```

```python
'''do1 > fo, do2 < fe; Virtual, inverted, enlarged final image --> microscope'''
case3 = geometrical_optics()
case3.add_lens(10, 20)
case3.add_lens(20, 15)
case3.ray_diagram(save_diagram = True)
```

```python
'''do1 > fo, do2 > fe; Real, upright, same size final image --> no microscope'''
case4 = geometrical_optics()
case4.add_lens(10, 20)
case4.add_lens(20, 40)
case4.ray_diagram(save_diagram = True)
```
